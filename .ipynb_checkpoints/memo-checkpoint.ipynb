```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
import dlib
```

```python
def reshape(src):
    map_x = np.zeros((src.shape[0], src.shape[1]), dtype=np.float32)
    map_y = np.zeros((src.shape[0], src.shape[1]), dtype=np.float32)
    h, w = map_x.shape[0], map_x.shape[1]
    
    cx,cy = 250,170 # 250,170
    bulgeRadius = 20 # 20
    bulgeStrength = 1.08
    
    for y in range(map_x.shape[0]):
        for x in range(map_x.shape[1]):
            dx, dy = x-cx, y-cy
            distanceSquared = dx * dx + dy * dy
            rx, ry = 1.0, 1.0
            
            if distanceSquared < bulgeRadius * bulgeRadius:
                distance = math.sqrt(distanceSquared)                
                rx = ((distance / bulgeRadius) ** 0.2 if distance>0 else 0) ** bulgeStrength
                ry = ((distance / bulgeRadius) ** 0.2 if distance>0 else 0) ** bulgeStrength
                
            map_x[y,x] = rx*dx + cx
            map_y[y,x] = ry*dy + cy

    return cv2.remap(src, map_x, map_y, cv2.INTER_AREA)

plt.rcParams['figure.figsize'] = (15.0, 10.0)

plt.subplot(121)
plt.imshow(cv2.cvtColor(src,cv2.COLOR_BGR2RGB))
plt.title("before")

plt.subplot(122)
plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.title("after")
```
