```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
def update_map(map_x, map_y):
    for y in range(map_x.shape[0]):
        for x in range(map_x.shape[1]):
            map_x[y,x] = x
            map_y[y,x] = y
```

```python
src = cv2.imread("/Users/yuheltanaka/kasuga.jpg")
```

```python
map_x = np.zeros((src.shape[0], src.shape[1]), dtype=np.float32)
map_y = np.zeros((src.shape[0], src.shape[1]), dtype=np.float32)
```

```python
update_map(map_x, map_y)
```

```python
dst = cv2.remap(src, map_x, map_y, cv2.INTER_LINEAR)
```

```python
plt.imshow(dst)
```

```python
cv2.imwrite("/Users/yuheltanaka/hoge.jpg",dst)
```
